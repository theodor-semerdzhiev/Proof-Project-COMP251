```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('HeapSort_out.csv')
n_data = np.arange(1,99999)
n_data = [x*1000 for x in n_data]
```

```python
n_vals = np.arange(1,99999)
n_logn_vals = (1/150000)*n_vals*np.log2(n_vals)
len(n_logn_vals)
```

```python
fig, ax = plt.subplots()
ax.plot(n_vals, n_logn_vals, linewidth=2)
ax.scatter(n_data, data, marker='o', c='orange', s=0.4)
# plt.ylim(0,20000)
# plt.xlim(10,10**4)
plt.title('Time Complexity of HeapSort')
# plt.xscale('log')

plt.tight_layout()
```
